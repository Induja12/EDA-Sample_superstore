```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("\nMissing values in the dataset:")
print(superstore_data.isnull().sum())
```

```python
# Calculate profit margin as a percentage of profit to sales
superstore_data['Profit Margin'] = (superstore_data['Profit'] / superstore_data['Sales']) * 100
```

```python
superstore_data['Profit Margin']
```

```python
# Analyze profit margins by Category
category_profit_margin = superstore_data.groupby('Category')['Profit Margin'].mean().reset_index()
```

```python
# Sort categories by profit margin
category_profit_margin = category_profit_margin.sort_values(by='Profit Margin')
```

```python
# Analyze profit margins by Sub-Category
sub_category_profit_margin = superstore_data.groupby('Sub-Category')['Profit Margin'].mean().reset_index()
```

```python
# Sort sub-categories by profit margin
sub_category_profit_margin = sub_category_profit_margin.sort_values(by='Profit Margin')
```

```python
# Profit margin by Category
plt.figure(figsize=(12, 6))
sns.barplot(data=category_profit_margin, x='Profit Margin', y='Category')
plt.title("Profit Margin by Category")
plt.xlabel("Profit Margin (%)")
plt.ylabel("Category")
plt.show()
```

```python
# Profit margin by Sub-Category
plt.figure(figsize=(12, 6))
sns.barplot(data=sub_category_profit_margin, x='Profit Margin', y='Sub-Category')
plt.title("Profit Margin by Sub-Category")
plt.xlabel("Profit Margin (%)")
plt.ylabel("Sub-Category")
plt.show()
```
